### numpy about creation
- `np.array` 创建数组
- `dtype` 矩阵数据类型
- `[[],[]]` 2维
- `np.zeros((n,m))` n*m的全零矩阵
- `np.ones((n,m))` n*m的全1矩阵
- `np.empty((n,m))` n*m的空矩阵，速度最快
- `np.arange(sum).reshape((n,m))` 从0-sum的n*m矩阵
- `np.linspace(n,m,s)` 从n到m一共s段的数组，reshape后也可使其分段
