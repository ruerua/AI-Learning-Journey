# NumPy 数组创建练习

```python
import numpy as np

# 创建一个一维数组
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)

# 创建一个全零数组
zeros = np.zeros((3, 3))
print(zeros)

# 创建一个随机数组
random_arr = np.random.randn(2, 3)
print(random_arr)
```

## 总结
今天学会了：
- `np.array()` 创建数组
- `np.zeros()` 创建全零矩阵
- `np.random.randn()` 创建随机数组
