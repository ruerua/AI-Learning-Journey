### 合并操作merge
- `pd.merge(left,right,on)` on = columns, 按照对应列表名进行合并，可以为['key1','key2'], 默认为how='inner'交集形式，进行连接;可以进行手动更改。除此之外，how可以等于'left'/'right',意为左连接/右连接
- `pd.merge()`里面还有参数suffixex['','']，里面给前面columns标题加后缀
