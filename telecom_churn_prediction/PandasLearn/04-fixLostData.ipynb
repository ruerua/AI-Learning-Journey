```python
import pandas as pd
import numpy as np
dates = pd.date_range('20130101', periods=6)
df = pd.DataFrame(np.arange(24).reshape(6,4),index=dates,columns=['A', 'B',
                                                                  'C', 'D'])
df.iloc[0,1] = np.nan
df.iloc[1,2] = np.nan
print(df)
print(df.dropna(axis=0,how='all'))
print(df.dropna(axis=0,how='any'))
print('##############################################')
print(df.dropna(axis=1,how='any'))
print('1111111111111111111111111111111111111111111111')
print(df,'\n',df.fillna(value=0))
print(df.isnull())
print(np.any(df.isnull()))
print(np.all(df.isnull()))
```

### 处理丢失数据
- `df.dropna(axis=0/1, how='any'/'all')` axis=0,它会按列顺序向下寻找，对于any,找到一个零，就把对应行删掉，how = all,则需要所有都是0。axis=1,则会按行顺序向右找，any,则找到一个零，就把对应列给删掉。
- `df.fillna(value=xxx)` 把df中所有nan变成value
- `df.isnull()` 矩阵元素变为True或False,如果原来是nan,则对应位置为True,否则为False
- `np.any(df.isnull())` 矩阵中只要有一个nan则返回True
