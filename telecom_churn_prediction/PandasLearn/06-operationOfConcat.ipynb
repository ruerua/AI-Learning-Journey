```python
import pandas as pd
import numpy as np
'''
df1 = pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1,columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4))*2,columns=['a','b','c','d'])
#print(df1)
#print(df2)
#print(df3)
res = pd.concat([df1,df2,df3],axis=0); print(res)
res1 = pd.concat([df1,df2,df3],axis=0, ignore_index=True); print(res1)
'''
#join
df1 = pd.DataFrame(np.ones((3,4))*0, columns=['a','b','c','d'],index=[1,2,3])
df2 = pd.DataFrame(np.ones((3,4))*1, columns=['b','c','d','e'],index=[2,3,4])
print(df1); print(df2)
res = pd.concat([df1,df2]); print(res) #新加列下会填充nan
res1 = pd.concat([df1,df2], join='outer',ignore_index=True); print(res1)
res2 = pd.concat([df1,df2], join='inner',ignore_index=True); print(res2)

```

### concat连接操作
- `pd.concat([], axis=0/1)` axis=0,竖向合并，axis=1,横向合并
- `pd.concat([], join='inner'/'outer')` inner为求交集，outer为求并集
