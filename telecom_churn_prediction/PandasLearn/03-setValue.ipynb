```python
import pandas as pd
import numpy as np
dates = pd.date_range('20130101', periods=6)
df = pd.DataFrame(np.arange(24).reshape((6,4)),
                  index = dates,
                  columns = ['A','B','C','D'])
print(df)
'''
df.iloc[2,2] = 1111
print(df)
df.loc['20130101', 'C'] = 2222
print(df)
'''
df[df.A > 4] = 0
print(df)
print('#############################')
df['F'] = np.nan #可通过这种方法加一列columns = 'F'的
print(df)
df['E'] = pd.Series(np.arange(1,7),
                    index = pd.date_range('20130101',periods=6))
print(df)
```

### 设置值
- `df.loc[] or df.iloc[]定位` 然后用=进行对应位置的赋值
- `df[df.A > 4] = 0` 把df中columns = 'A' 的所有行内数据赋为0
- `df.A[df.A > 4]` 把df中columns = 'A' 的 'A'列下的数据赋为0， u ca also use df.B[df.A > 4]
- `df['F'] = np.nan` 可通过这种方式对df添加一列columns = 'F'的
