```python
import pandas as pd
import numpy as np
dates = pd.date_range('20130101', periods=6)
df = pd.DataFrame(np.arange(24).reshape((6,4)),
                  index=dates,
                  columns=['A','B','C','D'])
print(df)
print(df['A'],df.A)
print('#################################################')
print(df[0:3],'\n',df['20130101':'20130103'])
print('1111111111111111111111111111111111111111111111111')
#selected by label: loc <- index
print(df.loc['20130102'])
print(df.loc['20130102', ['A','B']])
print('2222222222222222222222222222222222222222222222222')
#selected by label <- i of index
print(df.iloc[1:3,0:4])
print('333333333333333333333333333333333333333333333333')
print(df)
print(df[df.A > 8])
```

### 数据选择
- `df.date_range(date, periods = num)`以date开始，一共num天
- `df.DataFrame(dataMatrix, index, columns)`
- `df.loc[index]`以index为关键词进行数据选择
- `df.loc[index, columns]`以index为关键词，显示columns列表里面对应列标题下的数据
- `df.iloc[IofIndex, IofColumns]`以索引进行选择,我们也可以进行切片操作,df.iloc[1:3, 2:3]. And we can also df.iloc[[1,3,5], 0:4]对IofIndex进行不连续的切片
- `df[df.A > 8]` 筛选
