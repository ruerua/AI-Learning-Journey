```python
import pandas as pd
import numpy as np
s = pd.Series([1,3,6, np.nan, 44, 1])
print(s)
print('####################################')
dates = pd.date_range('20160101', periods = 6)
print(dates)
print('####################################')
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns = ['a','b','c','d'])
print(df)
print('1111111111111111111111111111111111111111')
df1 = pd.DataFrame(np.arange(12).reshape((3,4)))
print(df1)
print('2222222222222222222222222222222222222222')
df2 = pd.DataFrame({'A': 1., 
                    'B': pd.Timestamp('20130102'),
                    'C':pd.Series(1,index=list(range(4)),dtype='float32'),
                    'E': ['test','train','test','train']})
print(df2)
print(df2.dtypes)
print('333333333333333333333333333333333333333')
print(df2.sort_index(axis = 1, ascending = False))
print(df2.sort_index(axis = 0, ascending = False))
print('4444444444444444444444444444444444444444444')
print(df2.sort_values(by = 'E'))
```

### basic operation
- `pd.series` 一维数组
- `pd.date_range(chr, periods=m)` 从chr日期开始，连续m天
- `pd.DataFrame(data, index, columns)` index索引（最左边的列）,columns列标签（最上面一行）,np.random.randn()从正态分布中提取数据，6,4 -> 六行四列
- `pd.DataFrame({'A': , 'B':})`通过字典进行创建
- `pd.sort_index(axis = 0/1, ascending = True/False)` axis = 0对index排序，1以columns为基础进行排序。ascending = True正序，False倒序
- `pd.sort_values(by = columns)`可以通过by = 列标题，然后以对应列的值为基础，进行排序
